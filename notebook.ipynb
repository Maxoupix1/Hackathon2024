```python
import pandas as pd
data_left = pd.read_csv('left_accs_preproc.csv')
data_left.tail()
```

```python
data_right = pd.read_csv('right_accs_preproc.csv')
data_right.tail()
```

```python
# Count the number of rows for each day
data_right['time'] = pd.to_datetime(data_right['time'])
print(data_right['time'].dt.date.value_counts().sort_index())

data_left['time'] = pd.to_datetime(data_left['time'])
print(data_left['time'].dt.date.value_counts().sort_index())
```

```python
# Count the number of rows in day one for each hour
print(data_right[data_right['time'].dt.date == pd.to_datetime('2024-07-20').date()]['time'].dt.hour.value_counts().sort_index())
print(data_left[data_left['time'].dt.date == pd.to_datetime('2024-07-20').date()]['time'].dt.hour.value_counts().sort_index())
```

```python
# Concatenate the two dataframes using the datetime column of the right dataframe and using linear interpolation to join the left dataframe
data = pd.merge_asof(data_right, data_left, on='time', direction='nearest')
data.tail()
```

```python
# Set time columns as index
data_right_timeindex = data_right.set_index('time', inplace=False)
data_right_timeindex.rename(columns={'acc_x': 'acc_x_right', 'acc_y': 'acc_y_right', 'acc_z': 'acc_z_right'}, inplace=True)
data_left_timeindex = data_left.set_index('time', inplace=False)
data_left_timeindex.rename(columns={'acc_x': 'acc_x_left', 'acc_y': 'acc_y_left', 'acc_z': 'acc_z_left'}, inplace=True)

# Interpolate the left-hand data to match the right-hand time index
left_hand_interpolated = data_left_timeindex.reindex(data_right_timeindex.index).interpolate(method='time')

# Merge the two dataframes
merged_data = pd.concat([data_right_timeindex, left_hand_interpolated], axis=1)

# Reset the index (optional)
merged_data.reset_index(inplace=True)

print(merged_data)
```
